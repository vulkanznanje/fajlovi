```python
import matplotlib.pyplot as plt
```

```python
Kontinenti = ["Azija","Afrika","Evropa","Severna Amerika",
             "Južna Amerika","Australija i Okeanija","Antarktik"]
Površina = [30.3, 43.6, 10.4, 25.3, 17.6, 8.9, 13.6]
```

```python
plt.pie(Površina, labels=Kontinenti)
plt.show()
plt.close()
```

**DODAVANJE LEGENDE**

```python
plt.pie(Površina)
#funkcija za dodavanje legende:
plt.legend(Kontinenti, bbox_to_anchor = (1,1))
plt.show()
plt.close()
```
