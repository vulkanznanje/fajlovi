### Lekcija 3.7

Prvi zadatak
1. Sa našeg *GitHub* repozitorijuma na adresi https://github.com/vulkanznanje/Fajlovi uvezi u svesku tabelu iz *CSV* fajla *Mesečna potrošnja struje*, sačuvaj je u promenljivu *MesečnaPotrošnja* i prikaži njen sadržaj.
2. U tabelu dodaj 3 nove kolone kao na slici. Pomoću odgovarajućih formula izračunaj Prosečno vreme rada datih uređaja u toku meseca, potrošnju električne energije na mesečnom nivou i cenu u dinarima. Cena struje je 3.72 din/kwh.
3. Odredi najmanju vrednost u koloni Snaga [w], najveću u koloni Potrošnja [kwh] i ukupnu u koloni Cena [RSD].
